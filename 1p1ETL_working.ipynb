## 1.1 Working on missing data

At first, when we get the raw data, they are usually excel files, csv files or MySQL tables. Here we focus on excel files. Csv files are similar. MySQL tables are treated just like what we do after we finish ETL process and build up the data warehouse.

Raw data and real data (not simulated data) usually contain missing data, here we first try to detect the missing data. After we put the raw data into python pandas dataframe, the missing data will be denotated as np.NaN, which can be detected by isnull(). np.NaN is a float number, which occupies less space than None (None is an object).

From a numpy ndarray to a list:  
num = np.sum(df.isnull().values, axis = 1).tolist()

Get unique values of the list and sort the list:  
l1 = list(set(num))
l1.sort()

list comprehension and put the results in a dict:  
fre = {i:num.count(i) for i in l1}  
count() is a method in the list class

Row counts for 'Number of NaNs', construct a dataframe from a dict:  
pd.DataFrame(fre.items(), columns=['Number of NaNs', 'Count'])  
Row index is row numbers.

```python
import numpy as np

# np.ptp(np.sum(df.isnull().values, axis = 1), axis=0)
print [min(np.sum(df.isnull().values, axis = 1)), max(np.sum(df.isnull().values, axis = 1))]
# df.isnull().values is an array, np.sum(,axis=1) means sum the values of each row
print

num = np.sum(df.isnull().values, axis = 1).tolist()
# from a 1D numpy array to a list

l1 = list(set(num))
l1.sort()
# get unique values of a list and sort

fre = {i:num.count(i) for i in l1}
# list comprehension to replace for loop
# count is a method of list
# the result of list comprehension can be put in a dict
# print fre
# print

dt = pd.DataFrame(fre.items(), columns=['Number of NaNs', 'Count'])
# dict.items() can convert a dict to a 2D list
# or list(fre.items()) in python 3

print dt

ts = pd.DataFrame(np.sum(df.isnull().values, axis = 1))
print type(ts)
# ts is a 1D dataframe
# print ts
print ts[0].value_counts()
print type(ts[0].value_counts())
# series
print ts[0].value_counts().name
ts2 = pd.Series(np.sum(df.isnull().values, axis = 1))
print type(ts2)
# series
print ts2.value_counts()
print ts2.value_counts().dtype
print ts2.value_counts().name
```
